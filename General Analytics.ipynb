```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance_from_origin(self):
        return ((self.x ** 2) + (self.y ** 2)) ** 0.5
```

```python
dix = pd.read_csv('dixie.csv')
dix.head()
```

```python
copy_dix = dix.copy()
gpd_dix = gpd.GeoDataFrame(copy_dix, geometry=gpd.points_from_xy(copy_dix['longitude'], copy_dix['latitude']))
gpd_dix = gpd_dix.drop(columns=['Unnamed: 0','location', 'point'])
gpd_dix.set_crs(epsg=3857, inplace=True)
gpd_dix = gpd_dix.rename(columns={'latitude':'LATITUDE', 'longitude':'LONGITUDE'})
gpd_dix.head()
```

```python
gpd_fire.set_crs(epsg=3857, inplace=True)
gpd_dix_fire = gpd.sjoin_nearest(gpd_dix, gpd_fire, how='inner')
gpd_dix_fire = gpd_dix_fire.drop(columns=['index_right'])
gpd_dix_fire.head()
```

```python
perim = gpd.read_file('New folder/Dixie_Perimeter_8_26_21.shp')
perimeter = perim.explode(index_parts=True)
perimeter = perimeter.rename(columns={'geometry':'perimeter'})

perimeter_copy = perimeter.copy()
perimeter_copy['geometry'] = perimeter_copy['perimeter'].centroid
perimeter_copy = perimeter_copy.drop(columns=['OBJECTID', 'IncidentNa', 'FeatureCat', 'MapMethod', 'Comments',
       'DeleteThis', 'Label', 'FeatureAcc', 'FeatureSta',
       'IsVisible', 'CreateName', 'CreateDate', 'EditName', 'DateCurren',
       'PolygonDat', 'IRWINID', 'GlobalID', 'Acres_Auto', 'GISS_Misc'])
perimeter_copy.set_crs(epsg=3857, inplace=True, allow_override=True)
perimeter_copy.head()
```

```python
dixie_final = gpd_dix_fire.sjoin_nearest(perimeter_copy, how='inner')
dixie_final.head()
```
